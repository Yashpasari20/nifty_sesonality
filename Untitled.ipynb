```python
#calculating correlation
df = pd.concat([weekly_monday['Returns'],weekly_tuesday['Returns'],weekly_wednesday['Returns'],weekly_thursday['Returns'],weekly_friday['Returns']],axis=1,ignore_index=True)
```

```python
df.head()
```

```python
df[1] = df[1].shift(-1)
df[2] = df[2].shift(-2)
df[3] = df[3].shift(-3)
df[4] = df[4].shift(-4)
```

```python
df = df.dropna()
df = df.rename(index=str, columns={0: "monday", 1: "tuesday",2:"wednesday",3:"thursday",4:"friday"})
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df,figsize=(8,8),hist_kwds={"bins":50},alpha=0.1);
```

```python
import seaborn as sns
corr = df.corr()

corr.style.background_gradient(cmap='coolwarm')
```

```python
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```
